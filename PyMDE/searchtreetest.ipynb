```python
import torch
from pygel3d import hmesh, graph, jupyter_display as jd
import numpy as np
import pymde

limit = 5000

mnist = pymde.datasets.MNIST()
tensor = torch.load("tensor.pt")[:limit]
nparr = tensor.numpy()
```

```python
pymde.plot(tensor, 
           color_by=mnist.attributes['digits'][:limit],
        #    marker_size=20
           )
```

```python
m = hmesh.Manifold().from_points(nparr)
m.cleanup()
# jd.display(m)
```

```python
g = graph.from_mesh(m)
g.cleanup()
# jd.display(g)
```
