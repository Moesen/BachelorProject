```python
from collections import Counter
dct = {0: [(1, 2), (2, 3), (3,4)]}
count = Counter(dct)
print(count.most_common())
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([[1, 2], [3, 2]])

plt.scatter(a[:, 0], a[:, 1])
b = np.average(a, axis=0)
plt.scatter(b[0], b[1])
```
